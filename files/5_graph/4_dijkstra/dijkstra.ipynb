## Calcul des distances

```python
import heapq

def dijkstra(G, s): # G : graphe représenté par matrice d'adjacence pondérée
    dist = [float("inf")]*len(G) # dist[v] va contenir la distance de s vers v
    q = []
    heapq.heappush(q, (0, s))
    while len(q) > 0:
        d, u = heapq.heappop(q)
        if dist[u] == float("inf"):
            dist[u] = d
            for v in range(len(G)):
                if G[u][v] != float("inf"):
                    heapq.heappush(q, (dist[u] + G[u][v], v))
    return dist

dijkstra(G, 0)
```
